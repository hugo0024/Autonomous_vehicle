Imports:

```python
import os
import fnmatch
import numpy as np
import pandas as pd
from natsort import natsorted
from google.colab import files
from google.colab.patches import cv2_imshow
from google.colab import drive
```

Load in the training csv:

```python
drive.mount('drive')
df = pd.read_csv('/content/drive/MyDrive/MLiS2/training_norm.csv')
df.describe()
```

Check for missing data:

```python
a = df['image_id'].tolist()
b = [x for x in range(a[0], a[-1] + 1)]
missing_vals = list(set(a) ^ set(b))
print('Missing data:', missing_vals)
```

Load the training images and remove the corrupted images:

```python
images_dir = '/content/drive/MyDrive/MLiS2/training_data/training_data'
file_list = os.listdir(images_dir)
image_paths = []
pattern = "*.png"

for filename in file_list:
    if fnmatch.fnmatch(filename, pattern):
        image_paths.append(os.path.join(images_dir,filename))

image_paths = natsorted(image_paths)
image_paths = [string for string in image_paths if string != '/content/drive/MyDrive/MLiS2/training_data/training_data/3141.png']
image_paths = [string for string in image_paths if string != '/content/drive/MyDrive/MLiS2/training_data/training_data/3999.png']
image_paths = [string for string in image_paths if string != '/content/drive/MyDrive/MLiS2/training_data/training_data/4895.png']
image_paths = [string for string in image_paths if string != '/content/drive/MyDrive/MLiS2/training_data/training_data/8285.png']
image_paths = [string for string in image_paths if string != '/content/drive/MyDrive/MLiS2/training_data/training_data/10171.png']

images = pd.DataFrame (image_paths, columns = ['images'])
images.describe()
```

Combine the training data and images as a single df:

```python
df_data = pd.concat([df, images], axis=1)
df_data.head()
```
